```python
# Import necessary libraries for data analysis and visualization
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv("AndhraPradesh.csv")
data
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(data['Sno'], data['Installed Capacity (MTA)'])
plt.title('Scatter Plot')
plt.xlabel('Sno')
plt.ylabel('Installed Capacity (MTA)')
plt.show()
```

```python
# Bar Chart
plt.figure(figsize=(10, 6))
plt.bar(data['Sno'], data['Installed Capacity (MTA)'])
plt.title('Bar Chart')
plt.xlabel('Sno')
plt.ylabel('Installed Capacity (MTA)')
plt.show()
```

```python
# Histogram
plt.figure(figsize=(10, 6))
plt.hist(data['Installed Capacity (MTA)'], bins=5)
plt.title('Histogram')
plt.xlabel('Installed Capacity (MTA)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Line Plot
plt.figure(figsize=(10, 6))
plt.plot(data['Sno'], data['Installed Capacity (MTA)'])
plt.title('Line Plot')
plt.xlabel('Sno')
plt.ylabel('Installed Capacity (MTA)')
plt.show()
```

```python
# Vertical Bar Chart
plt.figure(figsize=(10, 6))
plt.barh(data['Name/Address'], data['Installed Capacity (MTA)'])
plt.title('Vertical Bar Chart')
plt.xlabel('Installed Capacity (MTA)')
plt.ylabel('Name/Address')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=data, x='Sno', y='Installed Capacity (MTA)')
plt.title('Seaborn Line Plot')
plt.show()
```
